```python
import os
import glob
import json
import random
import ipyplot
import mojimoji
from collections import defaultdict

from PIL import Image
```

```python
base_dir = 'denoised'
json_files = sorted(glob.glob(os.path.join(base_dir, '*.json')))
```

```python
def show_list(etl, query):
    label2file = defaultdict(list)

    for e in etl:
        label2file[e['label']].append(e['filename'])

    key = ord(mojimoji.han_to_zen(query))
    files = label2file[key]

    sampled = random.sample(files, 20)

    images = [Image.open(os.path.join(base_dir, f)) for f in sampled]
    ipyplot.plot_images(images, img_width=150)
```

```python
with open(json_files[0], 'r') as f:
    etl = json.load(f)

etl = list(filter(lambda x: x['valid'], etl))
show_list(etl, 'D')
```
